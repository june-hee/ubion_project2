```python
import pandas as pd

df = pd.read_excel('./data/추가기업.xlsx')
df['시간'] = pd.to_datetime(df['시간'])
df = df.drop('번호', axis=True)
df = df.sort_values('시간')
df = df.set_index(keys='시간')
df = df.reset_index()
df = df.drop_duplicates(['회사명'], keep = 'first') #관리종목 중복 지정 시 가장 먼저 지정된 시점만 남기고 중복 데이터 삭제
df['시간'] = df['시간'].dt.strftime("%Y.%m.%d")
df['시간'] = pd.to_datetime(df['시간'])
df = df.set_index(keys='시간')
df = df.reset_index()

df = df[['시간', '회사명', '종목코드', '공시제목', '제출인']]
df
```

```python
#관리종목 중복 지정 기업 확인
df[df['회사명']=='골드앤에스']
```

```python
ove = df[df.duplicated('회사명')] #중복 제거 확인
ove = ove.sort_values('회사명')
ove['회사명'].unique()
```

```python
from dateutil.relativedelta import relativedelta

#6개월 전 데이터 추출 위해 날짜 계산

new = []
for i in range(0, len(df)):
    n = df.loc[i]['시간'] - relativedelta(months=6)
    new.append(n)
```

```python
#데이터 프레임화
df['기간'] = pd.DataFrame(new)
```

```python
df = df[['시간', '회사명', '종목코드', '공시제목', '제출인', '기간']]
```

```python
df.info()
```

```python
#크롤링을 위한 날짜 형식 변경
df['시간'] = df['시간'].dt.strftime("%Y.%m.%d")
```

```python
df['기간'].astype('str')
df['기간'] = df['기간'].dt.strftime("%Y.%m.%d")
```
