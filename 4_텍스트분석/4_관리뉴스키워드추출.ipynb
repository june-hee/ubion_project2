```python
import pandas as pd

df = pd.read_csv('/Users/ijunhui/Downloads/ubion/프로젝트2/3_크롤링전처리/관리/플랫폼/최종_관리_플랫폼+닉변.csv')
df
```

```python
df['전체'] = df['제목'] + " " + df['본문']

df_gr_com = df.groupby('종목명').sum()
```

```python
import re

## () [] 안의 단어는 기자 이름 등 불필요한 단어이므로 제거하고
## 불필요한 단어 특수문자 제거
pattern1 = r'\([^)]*\)'
pattern2 = r"\[([^]]+)\]"

df_gr_com['뉴스기사본문전처리'] = [re.sub(pattern1, '', s) for s in df_gr_com['전체']]
df_gr_com['뉴스기사본문전처리'] = [re.sub(pattern2, '', s) for s in df_gr_com['뉴스기사본문전처리']]
df_gr_com['뉴스기사본문전처리'] = [re.sub('[^/^$|\s+/가-힣\.]', '', s) for s in df_gr_com['뉴스기사본문전처리']]
df_gr_com['뉴스기사본문전처리'] = [re.sub('[[ \s]{2,}\.{2,}]', '', s) for s in df_gr_com['뉴스기사본문전처리']]
df_gr_com['뉴스기사본문전처리'] = [re.sub('\n', '', s) for s in df_gr_com['뉴스기사본문전처리']]
df_gr_com['뉴스기사본문전처리'] = [re.sub('[/+]', '', s) for s in df_gr_com['뉴스기사본문전처리']]
df_gr_com['뉴스기사본문전처리'] = [re.sub('\r', '', s) for s in df_gr_com['뉴스기사본문전처리']]

df_gr_com['뉴스기사본문전처리']
df_gr_com
```

```python
korean_stopwors_path = "/Users/ijunhui/Downloads/ubion/프로젝트2/korean_stopword.txt"

# 불용어 텍스트 파일 오픈
with open (korean_stopwors_path, encoding= 'utf8') as f : 
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
```

```python

# 형태소 분석기 OKT를 사용한 토큰화 작업 (다소 시간 소요)

# 명사 단위 분류 -- 3기 명사 추출함

# 기업별 키워드 추출

okt = Okt()

tokenized_data = []
for sentence in (train_data):
    tokenized_sentence = okt.nouns(sentence) # 토큰화 # morphs 형태소 단위로 / .nouns()명사만 / phrases() 어절
    
    t_s = []
    for i in range(len(tokenized_sentence)):
        if len(tokenized_sentence[i]) > 1:
            t_s.append(tokenized_sentence[i])
        
    stopwords_removed_sentence = [word for word in t_s if not word in stopwords] # 불용어 제거
    tokenized_data.append(stopwords_removed_sentence)
```
