```python
# 구글 크롬 사용을 위해 webdriver 따로 다운로드 후 불러오기
from cgitb import html, text
from pydoc import cli
import click
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
from selenium.common.exceptions import NoSuchElementException

# chromedirver 불러오기
path = './chromedriver 2'
driver = webdriver.Chrome(path)
driver.maximize_window()
sleep(1)

com_l = []
title_l = []
good_l = []
bad_l = []
tur_l = []
chj_l = []
stall_l = []
star_l = []
day_l = []

Url = "https://www.teamblind.com/kr/"
driver.get(Url)
lin = driver.find_element(By.XPATH, "//*[@id='wrap']/header/div/div/div/a[2]").click()

comlist = ['아이에이'] #, '인피닉']
page = 10
sleep(15)

for i in comlist:
    for k in range(1, page):
        for j in range(1, 31):
            Url = "https://www.teamblind.com/kr/company/" + i + "/reviews?page=" + str(k)
            driver.get(Url)

            try:
                com = i

                day = driver.find_element(By.XPATH, "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div[1]/div/div[2]/div[2]").text
                day = day[-10:]

                stall = driver.find_element(By.XPATH, "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[1]/div/div[1]/div[1]/strong").text
                stall = stall[-3:]

                star = driver.find_element(By.XPATH, "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div[" + str(j) + "]/div/div[2]/div[1]/strong").text
                star = star[-3:]
                
                title = driver.find_element(By.XPATH, 
                                            "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div["+ str(j) + "]/div/div[2]/h3/a").text
                good = driver.find_element(By.XPATH, 
                                            "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div[" + str(j) + "]/div/div[2]/div[3]/div[1]/p[1]/span").text
                bad = driver.find_element(By.XPATH, 
                                            "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div[" + str(j) + "]/div/div[2]/div[3]/div[1]/p[2]/span").text
                try:
                    chj = driver.find_element(By.XPATH, "//*[@id='wrap']/section/div/div/div[2]/div/div/div/div/section[2]/div[2]/div[" + str(j) + "]/div/div[2]/div[3]/div[1]/p[3]/span").text
                    chj_l.append(chj)
                except:
                    chj_l.append("")
                    pass
                
                day_l.append(day)
                stall_l.append(stall)
                star_l.append(star)
                com_l.append(com)
                title_l.append(title)
                good_l.append(good)
                bad_l.append(bad)
                

            except:
                break
                

```
